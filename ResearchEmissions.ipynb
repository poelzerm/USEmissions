```python
#Reshape election data for merge 
election = Election_1996_2020_D.melt(id_vars=["STATE"], var_name="Year", value_name="Election_Result")
election['Year'] = election['Year'].astype(int)
election['Election_Result'] = election['Election_Result'].str.rstrip('%').astype(float)
election = election.rename(columns={'STATE':'State'})
election
```

```python
#Prepare data
emissions_per_state_clean = emissions_per_state.drop(columns={'2021', '2022', 'Percent', 'Absolute', 'Percent.1',	'Absolute.1'})
#Reshape emission data for merge 
emission = emissions_per_state_clean.melt(id_vars=["State"], var_name="Year", value_name="Energy-related carbon dioxide emissions")
emission['Year'] = emission['Year'].astype(int)

emission = emission.drop([2702,2701])
```

```python
#Prepare data
emissions_per_capita_clean = emissions_per_capita.drop(columns={'2021', '2022', 'Percent', 'Absolute', 'Percent.1',	'Absolute.1'})
#Reshape emission data for merge 
emission_pC = emissions_per_capita_clean.melt(id_vars=["State"], var_name="Year", value_name="Energy-related carbon dioxide emissions per Capita")
emission_pC['Year'] = emission_pC['Year'].astype(int)
emission_pC['Year'] = emission_pC['Year'].astype(int)
emission_pC = emission_pC.drop([2801,2802,2803,2804])
```

```python
#merge data
merged_data=pd.merge(election,emission, on=['State','Year'])
merged_data=pd.merge(merged_data,emission_pC, on=['State','Year'])
#ahjust from million metric tons 
merged_data['Energy-related carbon dioxide emissions'] = pd.to_numeric(merged_data['Energy-related carbon dioxide emissions'], errors='coerce')
merged_data['Energy-related carbon dioxide emissions'] = merged_data['Energy-related carbon dioxide emissions'] * 1000000
```

```python
#Pearson Correlation for total emissions and emissions per capita 
#Does not consider variations over the years or variations by state
from scipy.stats import pearsonr

democratic_percentage = merged_data['Election_Result']  # Democratic percentage (Election Result)
emissions = merged_data['Energy-related carbon dioxide emissions']  # Emissions in metric tons
emissions_per_capita = merged_data['Energy-related carbon dioxide emissions per Capita']  # Emissions per capita

# Calculate Pearson correlation between Democratic percentage and emissions
corr_emissions, p_value_emissions = pearsonr(democratic_percentage, emissions)

# Calculate Pearson correlation between Democratic percentage and emissions per capita
corr_emissions_per_capita, p_value_emissions_per_capita = pearsonr(democratic_percentage, emissions_per_capita)

# Print the results
print(f'Pearson correlation between Democratic percentage and emissions: {corr_emissions}')
print(f'P-value: {p_value_emissions}')

print(f'Pearson correlation between Democratic percentage and emissions per capita: {corr_emissions_per_capita}')
print(f'P-value: {p_value_emissions_per_capita}')

# Scatter plot for Democratic percentage vs emissions
plt.figure(figsize=(8, 6))
sns.scatterplot(x=democratic_percentage, y=emissions)
plt.title('Democratic Percentage vs Emissions')
plt.xlabel('Democratic Percentage (Election Result)')
plt.ylabel('Energy-related Carbon Dioxide Emissions (Metric Tons)')
plt.show()

# Scatter plot for Democratic percentage vs emissions per capita
plt.figure(figsize=(8, 6))
sns.scatterplot(x=democratic_percentage, y=emissions_per_capita)
plt.title('Democratic Percentage vs Emissions per Capita')
plt.xlabel('Democratic Percentage (Election Result)')
plt.ylabel('Energy-related Carbon Dioxide Emissions per Capita (Metric Tons)')
plt.show()
```

```python
# Calculate the correlation per year
years = merged_data['Year'].unique()

# Dictionary to store the correlations per year
correlations_by_year = {}

for year in years:
    year_data = merged_data[merged_data['Year'] == year]
    corr, _ = pearsonr(year_data['Election_Result'], year_data['Energy-related carbon dioxide emissions per Capita'])
    correlations_by_year[year] = corr

# Print correlations for each year
correlations_df = pd.DataFrame(list(correlations_by_year.items()), columns=['Year', 'Correlation'])


#create graph
x = correlations_df["Year"]
y = correlations_df["Correlation"]

plt.plot(x, y, marker='o')

plt.title('Correlation throughout the years')
plt.ylabel('Correlation')

# Show the plot
plt.show()
```

```python
# Calculate the correlation per state
states = merged_data['State'].unique()

# Correlations per state
correlations_by_state = {}

for state in states:
    state_data = merged_data[merged_data['State'] == state]
    corr, _ = pearsonr(state_data['Election_Result'], state_data['Energy-related carbon dioxide emissions per Capita'])
    correlations_by_state[state] = corr

# convert to data frame 
correlations_df2 = pd.DataFrame(list(correlations_by_state.items()), columns=['State', 'Correlation'])

#export data for vizualisation in Tableau
correlations_df2.to_excel("correlations_by_state.xlsx")
```

```python
# Calculate the average Democratic percentage (Election_Result) for each state
state_avg_democratic = merged_data.groupby('State')['Election_Result'].mean()

# Classify the states based on their average Democratic percentage
def classify_state_by_average_democratic(avg_democratic_percentage):
    if avg_democratic_percentage >= 62:
        return "Strongly Democratic leaning"
    elif 52 <= avg_democratic_percentage < 62:
        return "Moderatly Democratic leaning"  
    elif 48 <= avg_democratic_percentage < 52:
        return 'Swing State'
    elif 38 <= avg_democratic_percentage < 48:
        return 'Moderatly Republican leaning'
    else:
        return "Strongly Republican leaning"

state_classification = state_avg_democratic.apply(classify_state_by_average_democratic)

merged_data['State_Type'] = merged_data['State'].map(state_classification)


merged_data
```

```python
# Dictionary to store correlations by group
correlations_by_group = {"Strongly Democratic leaning" : [], "Moderatly Democratic leaning": [], 'Swing State': [], 'Moderatly Republican leaning': [], "Strongly Republican leaning": []}
# Loop over each group
for group in ["Strongly Democratic leaning", "Moderatly Democratic leaning", 'Swing State', 'Moderatly Republican leaning', "Strongly Republican leaning"]:
    group_data = merged_data[merged_data['State_Type'] == group]
    
    # Check if the group has data before calculating the correlation
    if len(group_data) > 0:
        corr = pearsonr(group_data['Election_Result'], group_data['Energy-related carbon dioxide emissions per Capita'])
        correlations_by_group[group] = corr

# Convert the result into a DataFrame
correlations_df = pd.DataFrame(list(correlations_by_group.items()), columns=['Group', 'Correlation, P-Value'])

# Print the DataFrame to see the results
correlations_df
```
